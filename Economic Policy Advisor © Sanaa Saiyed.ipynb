```python
# prompt: write a programme that analyses the country's economy and suggests economic policies based upon the economical factors that the country are struggling with.

# This is a simplified example and would require significant expansion and refinement for real-world application.

def analyze_economy(gdp_growth, inflation, unemployment, trade_balance):
  """
  Analyzes key economic indicators and suggests potential policies.

  Args:
    gdp_growth: GDP growth rate (%).
    inflation: Inflation rate (%).
    unemployment: Unemployment rate (%).
    trade_balance: Trade balance (positive for surplus, negative for deficit).

  Returns:
    A list of suggested economic policies.
  """

  policies = []

  if gdp_growth < 2:
    policies.append("Consider expansionary fiscal policy (e.g., increased government spending, tax cuts)")
    policies.append("Implement monetary easing (e.g., lower interest rates)")

  if inflation > 3:
    policies.append("Tighten monetary policy (e.g., raise interest rates)")
    policies.append("Consider fiscal restraint (e.g., reduced government spending)")

  if unemployment > 6:
    policies.append("Invest in job training and education programs")
    policies.append("Consider infrastructure spending to create jobs")

  if trade_balance < 0:
    policies.append("Promote export industries")
    policies.append("Consider tariffs or import restrictions (with caution)")

  return policies

# Example usage
gdp_growth = float(input("Enter GDP growth rate (%): "))
inflation = float(input("Enter inflation rate (%): "))
unemployment = float(input("Enter unemployment rate (%): "))
trade_balance = float(input("Enter trade balance (positive for surplus, negative for deficit): "))

suggested_policies = analyze_economy(gdp_growth, inflation, unemployment, trade_balance)

print("\nSuggested Economic Policies:")
for policy in suggested_policies:
  print("- ", policy)
```
