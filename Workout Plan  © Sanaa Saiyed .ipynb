```python
# prompt: write a programme which suggests a nutrition plan and exercise workout based on your body type and lifestyle.

# Function to determine body type based on user input
def get_body_type():
  print("To determine your body type, please answer the following questions:")
  shoulder_width = input("Are your shoulders broad or narrow? (broad/narrow): ").lower()
  waist_size = input("Is your waist wide or narrow? (wide/narrow): ").lower()
  hip_size = input("Are your hips wide or narrow? (wide/narrow): ").lower()

  if shoulder_width == "broad" and waist_size == "narrow" and hip_size == "narrow":
    return "Ectomorph"
  elif shoulder_width == "broad" and waist_size == "wide" and hip_size == "wide":
    return "Mesomorph"
  elif shoulder_width == "narrow" and waist_size == "wide" and hip_size == "wide":
    return "Endomorph"
  else:
    return "Mixed"

# Function to get user's activity level
def get_activity_level():
  print("\nHow active are you?")
  print("1. Sedentary (little to no exercise)")
  print("2. Lightly active (light exercise/sports 1-3 days/week)")
  print("3. Moderately active (moderate exercise/sports 3-5 days/week)")
  print("4. Very active (hard exercise/sports 6-7 days a week)")
  print("5. Extra active (very hard exercise/sports & physical job or 2x training)")

  while True:
    try:
      activity_level = int(input("Enter your activity level (1-5): "))
      if 1 <= activity_level <= 5:
        return activity_level
      else:
        print("Invalid input. Please enter a number between 1 and 5.")
    except ValueError:
      print("Invalid input. Please enter a number.")

# Function to suggest nutrition plan based on body type
def suggest_nutrition_plan(body_type):
  print("\n--- Suggested Nutrition Plan ---")
  if body_type == "Ectomorph":
    print("Focus on calorie-dense foods to gain weight.")
    print("- High carbohydrates (whole grains, fruits, vegetables)")
    print("- Moderate protein (lean meats, fish, eggs, beans)")
    print("- Moderate healthy fats (nuts, seeds, avocado)")
  elif body_type == "Mesomorph":
    print("Maintain a balanced diet to support muscle growth and maintain weight.")
    print("- Balanced carbohydrates (whole grains, fruits, vegetables)")
    print("- High protein (lean meats, fish, eggs, beans)")
    print("- Moderate healthy fats (nuts, seeds, avocado)")
  elif body_type == "Endomorph":
    print("Focus on a lower-carb, higher-protein diet to manage weight.")
    print("- Lower carbohydrates (focus on complex carbs like vegetables)")
    print("- High protein (lean meats, fish, eggs, beans)")
    print("- Low healthy fats (nuts, seeds, avocado in moderation)")
  else:
    print("Consult a nutritionist for a personalized plan based on your mixed body type.")

# Function to suggest workout plan based on body type and activity level
def suggest_workout_plan(body_type, activity_level):
  print("\n--- Suggested Workout Plan ---")
  if body_type == "Ectomorph":
    print("Focus on strength training to build muscle mass.")
    print("- Compound exercises (squats, deadlifts, bench press)")
    print("- Fewer reps with heavier weights (8-12 reps)")
    print("- Shorter, more intense workouts")
    if activity_level >= 3:
      print("- Include some cardio for cardiovascular health (2-3 days/week)")
  elif body_type == "Mesomorph":
    print("Combine strength training and cardio for a balanced approach.")
    print("- Strength training with moderate reps and weight (10-15 reps)")
    print("- Variety of exercises (compound and isolation)")
    print("- Cardio for endurance and fat burning (3-4 days/week)")
  elif body_type == "Endomorph":
    print("Focus on cardio for weight management and strength training for muscle tone.")
    print("- Higher reps with lighter weights (12-15 reps)")
    print("- More cardio (4-5 days/week)")
    print("- Include HIIT (High-Intensity Interval Training) for efficient fat burning")
  else:
    print("Consult a fitness professional for a personalized plan based on your mixed body type.")

# Main program
if __name__ == "__main__":
  body_type = get_body_type()
  print("\nYour body type is likely:", body_type)

  activity_level = get_activity_level()

  suggest_nutrition_plan(body_type)
  suggest_workout_plan(body_type, activity_level)

```
