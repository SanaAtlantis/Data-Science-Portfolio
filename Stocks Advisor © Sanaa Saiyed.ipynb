```python
# prompt: write a dynamic programme which analyses the stock markets on London Stock Exchange and advises the trader on what shares to buy and sell.

# This is a simplified example and should not be used for real trading.
# You will need to gather real-time stock data and implement sophisticated
# algorithms for accurate predictions.

def analyze_stocks():
  # Replace with actual stock data retrieval
  stock_data = {
    "ABC": {
      "price": 100,
      "trend": "up"
    },
    "XYZ": {
      "price": 50,
      "trend": "down"
    }
  }

  recommendations = []
  for stock, data in stock_data.items():
    if data["trend"] == "up":
      recommendations.append(f"Buy {stock} at {data['price']}")
    elif data["trend"] == "down":
      recommendations.append(f"Sell {stock} at {data['price']}")

  return recommendations

recommendations = analyze_stocks()
for recommendation in recommendations:
  print(recommendation)
```
